```python
import json
import csv
```

```python
taxa_json = '../genome-properties/docs/release/hierarchy.json'
```

```python
with open(taxa_json, 'r') as json_file:
    hierarchy = json.load(json_file)
```

```python
properties = {}
```

```python
class GenomeProperty(object):
    def __init__(self, gen_prop_id, name, parent=None, childern=[]):
        self.gen_prop_id = gen_prop_id
        self.name = name
        self.parent = parent
        self.childern = childern
```

```python
def parse_genome_property_json(gen_prop_dict, current_json_node, parent_property=None):
    gen_prop_id = current_json_node['id']
    name = current_json_node['name']
    children_json_nodes = current_json_node['children']
    
    current_genome_property = GenomeProperty(gen_prop_id, name, parent=parent_property)
    
    child_genome_properties = []
    for child_json in children_json_nodes:
        child_genome_prop = parse_genome_property_json(gen_prop_dict, child_json, parent_property=current_genome_property)
        child_genome_properties.append(child_genome_prop)
            
    current_genome_property.childern = child_genome_properties
    gen_prop_dict[current_genome_property.gen_prop_id] = current_genome_property
    
    return current_genome_property
```

```python
top_level_property = parse_genome_property_json(properties, hierarchy)
```

```python
def build_parent_child_list(genome_property, parent_child_mapping_list):
    parent_id = genome_property.gen_prop_id
    for child in genome_property.childern:
        parent_child_mapping_list.append((parent_id, child.gen_prop_id))
        build_parent_child_list(child, parent_child_mapping_list)
```

```python
parent_child_table = []
build_parent_child_list(top_level_property,parent_child_table)
```

```python
parent_child_table
```

```python
with open('./gen_prop_mappings.tsv', 'w') as tsv_file:
    csv_writer = csv.writer(tsv_file, delimiter='\t')
    for row in parent_child_table:
        csv_writer.writerow(row)
```
