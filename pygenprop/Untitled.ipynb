```python
indentifier_rows = []
for property_identifier in results.properties:
    genome_property = tree[property_identifier]
    for step in genome_property.steps:
        for identifier in step.consortium_identifiers:
            indentifier_rows.append((genome_property.id, step.number, identifier))
step_signature_accession = pd.DataFrame(indentifier_rows, columns=['Property_Identifier','Step_Number','Signature_Accession'])
```
