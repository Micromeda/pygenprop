```python
with open('testing/test_constants/genomeProperties.txt') as genome_property_file:
    properties_tree = parse_genome_property_file(genome_property_file)
```

```python
with open('testing/test_constants/LONGFORM_REPORT_GENOME_PROPERTIES_CA_L442_64_1.txt') as long_form_one:
    with open('testing/test_constants/LONGFORM_REPORT__C_chlorochromatii_CaD3.txt') as long_form_two:
        long_form_one_assignments = parse_genome_property_longform_file(long_form_one)
        long_form_two_assignments = parse_genome_property_longform_file(long_form_two)
```

```python
property_assignment_table = []
step_assignment_table = []
for genome_property_id, assignment_data in long_form_one_assignments.items():
    if bool(assignment_data['partial']):
        result = 'PARTIAL'
    else:
        result = 'YES'
    
    property_assignment_table.append([genome_property_id, result])
    
    all_step_numbers = set(step.number for step in properties_tree[genome_property_id].steps)
    supported_step_numbers = set(assignment_data['supported_steps'])
    unsupported_steps_numbers = all_step_numbers - supported_step_numbers
    
    for step_number in supported_step_numbers:
        step_assignment_table.append([genome_property_id, step_number, True])
        
    for step_number in unsupported_steps_numbers:
        step_assignment_table.append([genome_property_id, step_number, False])
```

```python
propery_assignments = pd.DataFrame(property_assignment_table, columns=['genome_property_id','property_result'])
propery_assignments.set_index(['genome_property_id'], inplace=True)
```

```python
step_assignment_table = pd.DataFrame(step_assignment_table, columns=['genome_property_id','step_number','step_result'])
step_assignment_table.set_index(['genome_property_id', 'step_number'], inplace=True)
```
